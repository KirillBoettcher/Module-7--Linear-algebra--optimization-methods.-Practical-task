```python
import numpy as np
def f(x, y):
    return (x - 3) ** 2 + (y + 5) ** 2

def grad_f(x, y):
    grad_x = 2 * x - 6
    grad_y = 2 * y + 10
    return np.array([grad_x, grad_y])
x0 = np.array([5, 2])
x_old = x0
# Эти параметры можно менять и смотреть на результат
eps = 10e-8
learning_rate = 0.1
max_steps = 100
i = 0
while np.linalg.norm(grad_f(x_old[0], x_old[1])) > eps:
# Обновляем x
    x_new = x_old - learning_rate * grad_f(x_old[0], x_old[1])
    x_old = x_new
# Увеличиваем счетчик
    i += 1
# Проверяем, нужно ли выходить из цикла
    if i == max_steps: 
        break
    print(x_new, f(x_new[0], x_new[1]), i)
```

```python
from scipy.optimize import minimize

x0 = [5, 2]

result = minimize(lambda x: f(x[0], x[1]), (x0,))
print(result)
```

```python
import matplotlib.pyplot as plt 
import numpy as np 
  
feature_x = np.linspace( -500, 500, 20) 
feature_y = np.linspace( -500, 500, 20) 
  
# Creating 2-D grid of features 
[X, Y] = np.meshgrid(feature_x, feature_y) 
  
fig, ax = plt.subplots(1, 1) 
  
Z = (X - 3)**2 + (Y + 5)**2
  
# plots filled contour plot 
ax.contourf(X, Y, Z) 
  
ax.set_title('График линий уровня') 
ax.set_xlabel('feature_x') 
ax.set_ylabel('feature_y') 
  
plt.show() 
```
